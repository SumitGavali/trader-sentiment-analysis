```python
import pandas as pd
from datetime import datetime

# Sample data creation (replace with your actual data loading)
# trades = pd.read_csv("trades.csv")
# sentiments = pd.read_csv("fear_greed_index.csv")

# 1. Enhanced date cleaning with error handling
def clean_dates(df, date_col):
    try:
        # First try ISO format (most common)
        df[date_col] = pd.to_datetime(df[date_col], errors='coerce').dt.date
        
        # If that fails, try parsing dayfirst
        if df[date_col].isna().any():
            df[date_col] = pd.to_datetime(df[date_col], dayfirst=True, errors='coerce').dt.date
            
        # If still failing, try inferring format
        if df[date_col].isna().any():
            df[date_col] = pd.to_datetime(df[date_col], infer_datetime_format=True, errors='coerce').dt.date
            
        return df
    except Exception as e:
        print(f"Error cleaning dates: {e}")
        return df

# Apply cleaning
trades = clean_dates(trades, 'Timestamp IST')
sentiments = clean_dates(sentiments, 'date')

# 2. Verify data before merging
print("\n=== Data Verification ===")
print("Trades date sample (first 5):")
print(trades['Timestamp IST'].head())
print("\nSentiments date sample (first 5):")
print(sentiments['date'].head())

# 3. Check for NA dates
print("\nNA values in dates:")
print(f"Trades: {trades['Timestamp IST'].isna().sum()}")
print(f"Sentiments: {sentiments['date'].isna().sum()}")

# 4. Date range comparison
print("\nDate ranges:")
print(f"Trades: {trades['Timestamp IST'].min()} to {trades['Timestamp IST'].max()}")
print(f"Sentiments: {sentiments['date'].min()} to {sentiments['date'].max()}")

# 5. Enhanced merge with diagnostics
try:
    # Create date-only columns for merging
    trades['trade_date'] = trades['Timestamp IST']
    sentiments['sentiment_date'] = sentiments['date']
    
    merged = pd.merge(
        trades,
        sentiments,
        left_on='trade_date',
        right_on='sentiment_date',
        how='left',
        indicator=True
    )
    
    # Analyze merge results
    print("\n=== Merge Results ===")
    print(merged['_merge'].value_counts())
    
    if (merged['_merge'] == 'left_only').any():
        missing_dates = merged[merged['_merge'] == 'left_only']['trade_date'].unique()
        print(f"\n{len(missing_dates)} trading dates missing sentiment data")
        print("Sample missing dates:", missing_dates[:5])  # Show first 5
        
    # Filter to only successful merges if needed
    merged = merged[merged['_merge'] == 'both'].drop(columns=['_merge', 'sentiment_date'])
    
except Exception as e:
    print(f"\nMerge failed: {e}")
    # Fallback to alternative merge method
    trades['date_str'] = trades['Timestamp IST'].astype(str)
    sentiments['date_str'] = sentiments['date'].astype(str)
    
    merged = pd.merge(
        trades,
        sentiments,
        left_on='date_str',
        right_on='date_str',
        how='left'
    )

# 6. Final merged data check
if not merged.empty:
    print("\n=== Merged Data ===")
    print(f"Merged dataframe shape: {merged.shape}")
    print("Columns:", merged.columns.tolist())
    print("\nSample merged rows:")
    print(merged.head())
else:
    print("\nMerge resulted in empty dataframe. Please check date formats.")
```

```python
import pandas as pd

# Load data
trades = pd.read_csv("historical_data.csv")
sentiments = pd.read_csv("fear_greed_index.csv")

# 1. Standardize date columns - use only one date column from each dataset
trades_date_col = 'Timestamp IST'  # or 'date' if more complete
sentiments_date_col = 'date'  # or 'Timestamp IST' if more complete

# 2. Convert dates properly
def convert_dates(df, col):
    # Try multiple parsing approaches
    df[col] = pd.to_datetime(df[col], errors='coerce').dt.normalize()
    if df[col].isna().any():
        df[col] = pd.to_datetime(df[col], dayfirst=True, errors='coerce').dt.normalize()
    return df

trades = convert_dates(trades, trades_date_col)
sentiments = convert_dates(sentiments, sentiments_date_col)

# 3. Verify dates
print("\nDate verification:")
print("Trades date range:", trades[trades_date_col].min(), "to", trades[trades_date_col].max())
print("Sentiments date range:", sentiments[sentiments_date_col].min(), "to", sentiments[sentiments_date_col].max())

# 4. Merge with diagnostics
merged = pd.merge(
    trades,
    sentiments,
    left_on=trades[trades_date_col].dt.date,
    right_on=sentiments[sentiments_date_col].dt.date,
    how='left',
    indicator=True
)

# 5. Analyze results
print("\nMerge results:")
print(merged['_merge'].value_counts())

if (merged['_merge'] == 'left_only').any():
    missing = merged[merged['_merge'] == 'left_only'][trades_date_col].nunique()
    print(f"\n{missing} trading dates have no sentiment data")

# Keep only successful merges
merged = merged[merged['_merge'] == 'both'].drop(columns=['_merge'])

# 6. Final check
if not merged.empty:
    print("\nSuccessful merge!")
    print("Merged shape:", merged.shape)
    print("Columns:", merged.columns.tolist())
else:
    print("\nMerge failed - no overlapping dates found")
```

```python
print("\n=== Sentiment Analysis ===")
sentiment_stats = merged.groupby('classification')['Closed PnL'].agg(['sum', 'mean', 'count']).round(2)
print(sentiment_stats)
```

```python
# Best/Worst Performing Sentiment Days
best_days = merged.groupby('date')['Closed PnL'].sum().nlargest(5)
worst_days = merged.groupby('date')['Closed PnL'].sum().nsmallest(5)

print("\nTop 5 Best Performing Days:")
print(best_days)
print("\nTop 5 Worst Performing Days:")
print(worst_days)

# Sentiment Effectiveness by Trade Side
print("\nPerformance by Trade Side and Sentiment:")
side_sentiment = pd.pivot_table(merged, 
                               index='Side',
                               columns='classification',
                               values='Closed PnL',
                               aggfunc='mean')
print(side_sentiment)

# 5. Risk Analysis
print("\nRisk Analysis:")
risk_stats = merged.groupby('classification')['Closed PnL'].agg(['std', 'max', 'min'])
print(risk_stats)
```
